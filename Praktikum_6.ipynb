```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import math
import os
import glob
```

```python
# Membuat histogram citra
original = cv.imread('/content/drive/MyDrive/PCVK/lena.jpg')
height, width, depth = np.shape(original)
names = np.arange(256)

blue = [0]*256
green = [0]*256
red = [0]*256

for y in range(0,height):
  for x in range(0,width):
    blue[original[y][x][0]] += 1
    green[original[y][x][1]] += 1
    red[original[y][x][2]] += 1

names = np.arange(256)
fig, axs = plt.subplots(1, 3, figsize=[20,5], sharex=True, sharey=True)
fig.suptitle('Histogram RGB plot')
fig.text(0.05, 0.5, 'Jumlah Kemunculan', va='center', rotation='vertical')
fig.text(0.5, 0.04, 'Intensitas Warna', ha='center')
axs[0].bar(names, blue, color='blue')
axs[1].bar(names, green, color='green')
axs[2].bar(names, red, color='red')
```

```python
# Membuat  histogram citra menggunakan library "histogram"
colors = ("b", "g", "r")
channel_ids = (0, 1, 2)

plt.xlim([0,256])
for channel_id, c in zip(channel_ids, colors):
  histogram, bin_edges = np.histogram(
      original[:, :, channel_id], bins=256, range=(0,256)
  )
  plt.plot(bin_edges[0:-1], histogram, color=c)

  plt.xlabel("Intensitas Warna")
  plt.ylabel("Jumlah Kemunculan")

  plt.show()
```

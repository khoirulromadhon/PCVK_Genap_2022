```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#3
original = cv2.imread('/content/drive/MyDrive/PCVK/j.png')

kernel = np.ones((5,5), np.uint8)
dilation = cv2.dilate(original, kernel, iterations = 1)

plt.subplot(121), plt.imshow(original), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(dilation), plt.title('Citra Hasil Dilasi')
```

```python
#4
kernel = np.ones((5,5), np.uint8)
erosion = cv2.erode(original, kernel, iterations = 1)

plt.subplot(121), plt.imshow(original), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(erosion), plt.title('Citra Hasil Erosi')
```

```python
#5
kernel = np.ones((5,5), np.uint8)
dilation = cv2.morphologyEx(original, cv2.MORPH_DILATE, kernel)

plt.subplot(121), plt.imshow(original), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(dilation), plt.title('Citra Hasil Dilasi')
```

```python
#6
kernel = np.ones((5,5), np.uint8)
erosion = cv2.morphologyEx(original, cv2.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(original), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(erosion), plt.title('Citra Hasil Erosi')
```
